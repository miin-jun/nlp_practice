1. 긍정/ 부정 기준 만들기

```python
import numpy as np

positive_texts = [
    "정말 좋다 만족스럽다 추천한다 최고다",
    "품질이 좋고 매우 훌륭하다 만족한다",
    "아주 마음에 든다 좋다 편리하다"
]

negative_texts = [
    "정말 별로다 실망스럽다 불편하다 최악이다",
    "품질이 나쁘고 만족하지 않는다 불만이다",
    "다시는 사용하고 싶지 않다 문제 많다"
]

def sentence_vector(text, model):
    words = text.split()
    vectors = [model.wv[word] for word in words if word in model.wv]
    
    if len(vectors) == 0:
        return np.zeros(model.vector_size)
    
    return np.mean(vectors, axis=0)

positive_vectors = [sentence_vector(text, model) for text in positive_texts]
negative_vectors = [sentence_vector(text, model) for text in negative_texts]

positive_center = np.mean(positive_vectors, axis=0)
negative_center = np.mean(negative_vectors, axis=0)

print("긍정 중심 벡터 크기:", positive_center.shape)
print("부정 중심 벡터 크기:", negative_center.shape)
print("긍정 중심 벡터 앞 10개:", positive_center[:10])
print("부정 중심 벡터 앞 10개:", negative_center[:10])
```
