```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

positive_texts = [
    "정말 좋다 만족스럽다 추천한다 최고다",
    "품질이 좋고 매우 훌륭하다 만족한다",
    "아주 마음에 든다 좋다 편리하다",
]

negative_texts = [
    "정말 별로다 실망스럽다 불편하다 최악이다",
    "품질이 나쁘고 만족하지 않는다 불만이다",
    "다시는 사용하고 싶지 않다 문제 많다",
    "생각보다 별로다",
    "사용하기에 너무 불편하다"
]

input_text = "제품이 불편하고 너무 실망스러웠다"

train_texts = positive_texts + negative_texts
labels = [1] * len(positive_texts) + [0] * len(negative_texts)

vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(train_texts)
X_input = vectorizer.transform([input_text])

sims = cosine_similarity(X_input, X_train)[0]
print("유사도:", sims)

pos_sims = [sim for sim, label in zip(sims, labels) if label == 1]
neg_sims = [sim for sim, label in zip(sims, labels) if label == 0]

pos_score = np.mean(pos_sims)
neg_score = np.mean(neg_sims)

print("긍정 평균 유사도:", pos_score)
print("부정 평균 유사도:", neg_score)

if pos_score > neg_score:
    print("판정 결과: 긍정")
else:
    print("판정 결과: 부정")
```
